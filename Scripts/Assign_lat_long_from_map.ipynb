```python
import pandas as pd

# load your tables
metadata = pd.read_csv("/Volumes/Ifeanyi/SLE+2025-07-30/IPD/DTA_loc.csv")
location = pd.read_csv("/Volumes/Ifeanyi/SLE+2025-07-30/IPD/location_lat_long.csv")

# merge on the district name
merged = metadata.merge(
    location[['NAME_2','lat','long']],
    left_on  ='District',
    right_on ='NAME_2',
    how      ='left'
)

# if you don’t need NAME_2 any more, drop it
merged = merged.drop(columns=['NAME_2'])
merged.to_csv("/Volumes/Ifeanyi/SLE+2025-07-30/IPD/location+meta+lat_long.csv")
# now merged has columns: label, location, lat, long
print(merged.head())
```
